```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.loadtxt('test/resources/X.csv', delimiter=',')
y = np.loadtxt('test/resources/y.csv', delimiter=',')
```

```python
fig = plt.figure(figsize=[15,5])
plt.plot(X/10, 'r')
plt.plot(y[0], 'k')
plt.xlim([3500,5500])
```

```python
fig = plt.figure(figsize=[15,5])
plt.plot(X,y[0],'k.')
plt.xlim([-10,65]);
```

```python
rng = [np.nanpercentile(X,1), np.nanpercentile(X,99)]
bins = np.linspace(rng[0], rng[1], 10)
centers = bins[0:-1] + abs(bins[1]-bins[0])/2
```

```python
tuning = np.zeros(len(bins)-1)

for i in range(len(bins)-1):
    t = (X > bins[i]) * (X < bins[i+1])
    tuning[i] = np.nanmean(y[0][t])
```

```python
fig = plt.figure(figsize=[15,5])
plt.plot(centers, tuning, 'k')
plt.xlim([-10,65]);
```
